## casting

```python
# int to float

int_1 = 2
print(type(int_1))

float_1 = float(int_1) # casting
print(type(float_1))
```

```python
# string to int

str_1 = "12"
print(type(str_1))

int_1 = int(str_1) # casting
type(int_1)
```

```python
# int to str

int_1 = 12
str_1 = str(int_1) # casting
type(str_1)
```

```python
# try different data type operation

"12" + 13
```
